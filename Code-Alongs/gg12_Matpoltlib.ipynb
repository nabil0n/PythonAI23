# Matplotlib


```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [value**2 for value in x]

print(f"{x = }")
print(f"{y = }")



y2 = [value**3 for value in x]
plt.plot(x,y, label="x**2")
plt.plot(x,y2, label="x**3")
```

```python
x = [0, -100, 25, 65, -323]
y = [0, 3, 7, 3, 9]

#plt.axis([-40, 30, 4, 6])
plt.title("Min graf")
plt.xlabel("x label")
plt.ylabel("y label")
plt.plot(x, y, color="red", label="Hej", linestyle="dotted", marker=".")
```

```python
x = list(range(-10, 11))
y = [value**3 for value in x]

plt.scatter(x, y, marker="x")
```

```python
labels = ["Hane", "Hona"]
values = [15, 23]

plt.bar(labels, values)

```

```python
labels = ["Hane", "Hona"]
values = [15, 23]

plt.pie(values, labels = labels)
```

```python
x2 = [value**2 for value in x]
y2 = [value**3 for value in x]

fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[1].plot(x2, y2)
```
